```python
def classify_triangle(a, b, c):
    """
    Phân loại tam giác dựa trên độ dài 3 cạnh.
    
    Args:
        a, b, c: Độ dài 3 cạnh của tam giác
        
    Returns:
        str: "Invalid Input", "Not a Triangle", "Equilateral", "Isosceles", hoặc "Scalene"
    """
    # Kiểm tra tính hợp lệ của dữ liệu đầu vào
    if not isinstance(a, int) or not isinstance(b, int) or not isinstance(c, int):
        return "Invalid Input"
    
    if a < 1 or a > 100 or b < 1 or b > 100 or c < 1 or c > 100:
        return "Invalid Input"
    
    # Kiểm tra xem có phải là tam giác hợp lệ
    if not (a + b > c and a + c > b and b + c > a):
        return "Not a Triangle"
    
    # Phân loại tam giác
    if a == b and b == c:
        return "Equilateral"
    elif a == b or b == c or a == c:
        return "Isosceles"
    else:
        return "Scalene"
```

```python
# Test cases
test_cases = [
    # Invalid Input
    (0, 1, 1, "Invalid Input"),  # a = 0
    (1, 101, 1, "Invalid Input"),  # b > 100
    (-1, 5, 5, "Invalid Input"),  # a < 0
    (1.5, 2, 3, "Invalid Input"),  # float instead of int
    
    # Not a Triangle
    (1, 2, 5, "Not a Triangle"),  # 1+2 < 5
    (1, 1, 3, "Not a Triangle"),  # 1+1 < 3
    (2, 3, 10, "Not a Triangle"),  # 2+3 < 10
    
    # Equilateral
    (5, 5, 5, "Equilateral"),
    (10, 10, 10, "Equilateral"),
    (1, 1, 1, "Equilateral"),
    (100, 100, 100, "Equilateral"),
    
    # Isosceles
    (5, 5, 3, "Isosceles"),
    (5, 3, 5, "Isosceles"),
    (3, 5, 5, "Isosceles"),
    (2, 2, 3, "Isosceles"),
    
    # Scalene
    (3, 4, 5, "Scalene"),
    (5, 6, 7, "Scalene"),
    (2, 3, 4, "Scalene"),
    (10, 20, 25, "Scalene"),
]

# Chạy test
print("Testing Triangle Classification Module\n")
print("=" * 60)
passed = 0
failed = 0

for a, b, c, expected in test_cases:
    result = classify_triangle(a, b, c)
    status = "✓ PASS" if result == expected else "✗ FAIL"
    
    if result == expected:
        passed += 1
    else:
        failed += 1
    
    print(f"{status}: classify_triangle({a}, {b}, {c}) = {result} (expected: {expected})")

print("=" * 60)
print(f"Results: {passed} passed, {failed} failed out of {len(test_cases)} tests")
```
