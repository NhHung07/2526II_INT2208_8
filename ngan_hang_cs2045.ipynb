```python
def loan_decision(age, income, credit_score, employment):
    # Kiểm tra tính hợp lệ của dữ liệu đầu vào
    if not isinstance(age, int) or not (18 <= age <= 65):
        return "Invalid Input"
    if not isinstance(income, float) or not (5.0 <= income <= 500.0):
        return "Invalid Input"
    if not isinstance(credit_score, int) or not (300 <= credit_score <= 850):
        return "Invalid Input"
    if employment not in ["C", "F"]:
        return "Invalid Input"
    
    # Phân loại rủi ro
    if 300 <= credit_score <= 500:
        risk = "High"
    elif 501 <= credit_score <= 700:
        risk = "Medium"
    else:       
        risk = "Low"

    # Logic nghiệp vụ
    if risk == "High":
        return "REJECT"
    
    if income < 15.0:
        if employment == "F" or risk == "Medium":
            return "REJECT"
        elif employment == "C" and risk == "Low":
            return "MANUAL REVIEW"

    if income >= 15.0 and risk in ["Medium", "Low"]:
        if employment == "C":
            return "APPROVE"
        elif employment == "F":
            return "MANUAL REVIEW"
```

```python
# Test cases
test_cases = [
    # Nhóm 1: Invalid Inputs
    {"id": "TC_01", "args": (17, 50.0, 750, "C"), "expected": "Invalid Input"},
    {"id": "TC_02", "args": (66, 50.0, 750, "C"), "expected": "Invalid Input"},
    {"id": "TC_03", "args": (30, 4.9, 750, "C"), "expected": "Invalid Input"},
    {"id": "TC_04", "args": (30, 500.1, 750, "C"), "expected": "Invalid Input"},
    {"id": "TC_05", "args": (30, 50.0, 299, "C"), "expected": "Invalid Input"},
    {"id": "TC_06", "args": (30, 50.0, 851, "C"), "expected": "Invalid Input"},
    {"id": "TC_07", "args": (30, 50.0, 750, "A"), "expected": "Invalid Input"},
    
    # Nhóm 2: Logic nghiệp vụ
    {"id": "TC_08", "args": (30, 50.0, 500, "C"), "expected": "REJECT"},
    {"id": "TC_09", "args": (18, 50.0, 300, "C"), "expected": "REJECT"},
    {"id": "TC_10", "args": (30, 14.9, 600, "C"), "expected": "REJECT"},
    {"id": "TC_11", "args": (65, 10.0, 850, "F"), "expected": "REJECT"},
    {"id": "TC_12", "args": (30, 5.0, 701, "C"), "expected": "MANUAL REVIEW"},
    {"id": "TC_13", "args": (30, 15.0, 501, "C"), "expected": "APPROVE"},
    {"id": "TC_14", "args": (30, 50.0, 700, "F"), "expected": "MANUAL REVIEW"}
]

# Chạy test
print(f"{'TC_ID':<7} | {'KẾT QUẢ THỰC TẾ':<15} | {'TRẠNG THÁI'}")
print("-" * 40)
passed_count = 0

for tc in test_cases:
    actual_result = loan_decision(*tc["args"])
    status = "✅ PASS" if actual_result == tc["expected"] else f"❌ FAIL (Expected: {tc['expected']})"
    if "PASS" in status: passed_count += 1
    print(f"{tc['id']:<7} | {actual_result:<15} | {status}")

print("-" * 40)
print(f"Tổng kết: {passed_count}/{len(test_cases)} Test Cases passed.")
```
